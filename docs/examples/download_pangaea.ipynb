# Download data from PANGAEA

[PANGAEA](www.pangaea.de) is an Open Access library aimed at archiving, publishing and distributing georeferenced data from earth system research. The system guarantees long-term availability of its content through a commitment of the hosting institutions. PANGAEA is open to any project, institution, or individual scientist to use or to archive and publish data.

In this example, we will download and analyze the [data of a CTD](https://doi.pangaea.de/10.1594/PANGAEA.883836) with id: 883836, published by the [OBSEA](www.obsea.es) team during the [EGIM](http://www.emsodev.eu/Files/Brochure/EGIM%20Brochure.pdf) test of [EMSO](www.emso.eu).

First, we import as follow:

```python
from mooda.access import Pangaea
```

Now, we are going to download the data form PANGAEA and check the Data Quality Control.

```python
id_data = 883836  # Id of the data

# Download the data from Pangea and create a WaterFrame
wf = Pangea.from_source_to_waterframe(id_data)
# Make the QC Tests
wf.qc()
wf
```

Now, we can continue working with the WaterFrame as usual...

```python
wf.tsplot("Temp")
```

Save the WaterFrame into a CSV and NetCDF

```python
wf.to_csv("pangea.csv")
```

```python
wf.to_netcdf("pangea.nc")
```
